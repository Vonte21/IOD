````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    '''
    sumX=sum(X) # sum all X
    sumY=sum(Y) # sum all Y
    X2 = [X ** 2 for X in X] # Square each X by X
    Y2 = [Y ** 2 for Y in Y] # Square each Y by Y
    XY = [a * b for a, b in zip(X, Y)] #multiply X values with Y values
    
    N=(len(X)+len(Y))/2 #get Dof
    m=(N*sum(XY)-sumX*sumY)/(N*sum(X2)-sumX**2) # get m as per formula above
    b=(sumY-m*sumX)/N # get b as per formula above
    

    
    # To Do
    
    return m, b

# solution
m, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x
y = [X*m+b for X in X] # get y values using m,b

# # To Do: plot points and fit line

plt.plot(X,Y,'o')
plt.plot(X,y)
````
